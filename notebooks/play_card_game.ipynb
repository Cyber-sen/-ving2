### Oppgave 1: Lag en tekstbasert meny

Vi skal nå ta i bruke klassene PlayCard, DeckOfCards og HandOfCards til å spille kort.

I denne oppgaven, skal du lage en funksjon "meny()" som viser en tekstbasert meny for et kortspill. Menyen skal gi brukeren flere valg, som å dele ut en hånd, vise hånden, sjekke hånden for spesifikke kombinasjoner, og avslutte programmet. Programmet skal fortsette å vise menyen til brukeren velger å avslutte.
Husk at det er kun menyen vi skal lage i denne oppgaven og ikke funksjonalitet bak valgene.

Menyvalg:

- "1. Del hånd" - Del ut en ny hånd med 5 kort.
- "2. Vis hånd" - Vis de nåværende kortene i hånden.
- "3. Sjekk hånd" - Sjekk om hånden har en flush.

Eksempel på kjøring:

Nåværende hånd: ...

Hånden har en flush!

1. Del hånd

**Tip**
- Bruk en løkke for å vise menyen gjentatte ganger til brukeren velger å avslutte.

````python
import sys
import os

# Add the parent directory to sys.path
sys.path.insert(0, os.path.abspath(os.path.join(os.getcwd(), '..', 'src')))

# Skriv ditt svar her:
from DeckOfCards import DeckOfCards

def display_menu():
    print("Velkommen til kortspillet!")
    print("1. Del hånd")
    print("2. Vis hånd")
    print("3. Sjekk hånd")
    print("4. Avslutt")

def deal_hand(deck):
    return deck.deal_hand(5)

def show_hand(hand):
    if hand:
        print(f"Nåværende hånd: {hand}")
    else:
        print("Ingen hånd delt ut ennå.")

def check_hand(hand):
    if hand:
        if hand.is_flush():
            print("Hånden har en flush!")
        else:
            print("Hånden har ikke en flush.")
    else:
        print("Ingen hånd delt ut ennå.")

def main():
    deck = DeckOfCards()
    hand = None

    while True:
        display_menu()
        choice = input("Velg et alternativ: ")
        if choice == '1':
            hand = deal_hand(deck)
            print(f"Ny hånd delt ut: {hand}")
        elif choice == '2':
            show_hand(hand)
        elif choice == '3':
            check_hand(hand)
        elif choice == '4':
            print("Programmet avsluttes. Ha en fin dag!")
            break
        else:
            print("Ugyldig valg, prøv igjen.")

if __name__ == "__main__":
    main()
````

### Oppgave 2

I denne oppgaven skal du analysere kortene på hånd (se punktliste under). Dette kan skje når brukeren trykker på menyvalget "Check hand".

Her er det meningen at du skal utvide metoden check_hand samt å anvende streams med tilhørende funskjoner for filtrering, map, reduce osv. Prøv å løse så mange som du klarer av følgende:

- Regn ut summen av alle verdiene av kortene på hånd (ess = 1) 
- Hent ut bare kort som er av fargen "Hjerter", og vis i et tekstfelt på formen "H12 H9 H1". Dersom det ikke er noen Hjerter på hånd, kan tekstfeltet inneholde teksten "No Hearts", for eksempel.
- Sjekk om kortet "Spar dame" finnes blant kortene på hånden.


````python
# Skriv ditt svar her:
from functools import reduce

# Metoden check_hand i DeckOfCards.py erstattes med denne:
def check_hand(hand):
    if hand:
        # Regn ut summen av alle verdiene av kortene på hånd (ess = 1)
        total_value = reduce(lambda acc, card: acc + card.get_face(), hand.cards, 0)
        print(f"Summen av alle verdiene av kortene på hånd: {total_value}")

        # Hent ut bare kort som er av fargen "Hjerter"
        hearts = list(filter(lambda card: card.get_suit() == 'H', hand.cards))
        hearts_str = " ".join(map(lambda card: card.get_as_string(), hearts)) if hearts else "No Hearts"
        print(f"Hjerter kort på hånd: {hearts_str}")

        # Sjekk om kortet "Spar dame" finnes blant kortene på hånden
        spade_queen = any(map(lambda card: card.get_suit() == 'S' and card.get_face() == 12, hand.cards))
        print(f"Spar dame finnes på hånden: {'Ja' if spade_queen else 'Nei'}")
        
        # Sjekk om kortene på hånd utgjør en "5-flush"
        suits = ['H', 'D', 'C', 'S']
        is_5_flush = any(len(list(filter(lambda card: card.get_suit() == suit, hand.cards))) == 5 for suit in suits)
        print(f"Hånden utgjør en 5-flush: {'Ja' if is_5_flush else 'Nei'}")
        
    else:
        print("Ingen hånd delt ut ennå.")
        

# Del kort (hånd) til spilleren
deck = DeckOfCards()
hand = deck.deal_hand(5)

check_hand(hand)
````

### Oppgave 4 (Frivillig)

Utvid programmet slik at det er mulig å lagre HandOfCards på en fil. Det skal også være mulig å lese HandOfCards fra en fil inn i programmet igjen.

````python
# Skriv ditt svar her:

import sys
import os
import pickle # for å lagre/laste et objekt til/fra fil

# Add the parent directory to sys.path
sys.path.insert(0, os.path.abspath(os.path.join(os.getcwd(), '..', 'src')))

# Skriv ditt svar her:
from DeckOfCards import DeckOfCards

def display_menu():
    print("Velkommen til kortspillet!")
    print("1. Del hånd")
    print("2. Vis hånd")
    print("3. Sjekk hånd")
    print("4. Lagre hånd til fil")
    print("5. Last hånd fra fil")
    print("6. Avslutt")

def deal_hand(deck):
    return deck.deal_hand(5)

def show_hand(hand):
    if hand:
        print(f"Nåværende hånd: {hand}")
    else:
        print("Ingen hånd delt ut ennå.")

def check_hand(hand):
    if hand:
        if hand.is_flush():
            print("Hånden har en flush!")
        else:
            print("Hånden har ikke en flush.")
    else:
        print("Ingen hånd delt ut ennå.")

def save_hand_to_file(hand, filename='hand_of_cards.pkl'):
    if hand:
        with open(filename, 'wb') as file:
            pickle.dump(hand, file)
        print(f"Hånden er lagret til {filename}.")
    else:
        print("Ingen hånd å lagre.")

def load_hand_from_file(filename='hand_of_cards.pkl'):
    if os.path.exists(filename):
        with open(filename, 'rb') as file:
            hand = pickle.load(file)
        print(f"Hånden er lastet fra {filename}.")
        return hand
    else:
        print(f"Filen {filename} finnes ikke.")
        return None

def main():
    deck = DeckOfCards()
    hand = None

    while True:
        display_menu()
        choice = input("Velg et alternativ: ")
        if choice == '1':
            hand = deal_hand(deck)
            print(f"Ny hånd delt ut: {hand}")
        elif choice == '2':
            show_hand(hand)
        elif choice == '3':
            check_hand(hand)
        elif choice == '4':
            save_hand_to_file(hand)
        elif choice == '5':
            hand = load_hand_from_file()
            print(f"Nåværende hånd: {hand}")
        elif choice == '6':
            print("Programmet avsluttes. Ha en fin dag!")
            break
        else:
            print("Ugyldig valg, prøv igjen.")

if __name__ == "__main__":
    main()
````
